```python
def convert_units():
    """
    Convert between different units of measurement.
    Supports length, weight and volume conversions.
    """

    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = int(input("Enter choice: "))
        value = float(input("Enter value to convert: "))

        if choice == 1:
            print("1. meters to feet")
            print("2. feet to meters")
            sub = int(input("Select option: "))

            if sub == 1:
                result = value * 3.28084
                print("Feet:", result)
            elif sub == 2:
                result = value / 3.28084
                print("Meters:", result)

        elif choice == 2:
            print("1. kg to lbs")
            print("2. lbs to kg")
            sub = int(input("Select option: "))

            if sub == 1:
                print("Pounds:", value * 2.20462)
            elif sub == 2:
                print("Kilograms:", value / 2.20462)

        elif choice == 3:
            print("1. liters to gallons")
            print("2. gallons to liters")
            sub = int(input("Select option: "))

            if sub == 1:
                print("Gallons:", value * 0.264172)
            elif sub == 2:
                print("Liters:", value / 0.264172)

        else:
            print("Invalid choice")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

convert_units()
```

```python
def fsum(numbers):
    return sum(numbers)

def faverage(numbers):
    return sum(numbers) / len(numbers)

def fmax(numbers):
    return max(numbers)

def fmin(numbers):
    return min(numbers)

try:
    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    print("Choose operation:")
    print("1 Sum")
    print("2 Average")
    print("3 Maximum")
    print("4 Minimum")

    choice = int(input("Choice: "))

    if choice == 1:
        print("Sum:", fsum(nums))
    elif choice == 2:
        print("Average:", faverage(nums))
    elif choice == 3:
        print("Max:", fmax(nums))
    elif choice == 4:
        print("Min:", fmin(nums))

except:
    print("Invalid input")
```

```python
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other([1,2,3,4,5,6]))
```

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]

print(get_sublist([10,20,30,40,50],1,3))
```

```python
def sum_nested(lst):
    total = 0
    for i in lst:
        if isinstance(i, list):
            total += sum_nested(i)
        else:
            total += i
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
def remove_element(lst, elem):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.append(remove_element(i, elem))
        elif i != elem:
            result.append(i)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
def find_max(lst):
    maximum = float('-inf')

    for i in lst:
        if isinstance(i, list):
            maximum = max(maximum, find_max(i))
        else:
            maximum = max(maximum, i)

    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

```python
def count_occurrences(lst, elem):
    count = 0

    for i in lst:
        if isinstance(i, list):
            count += count_occurrences(i, elem)
        elif i == elem:
            count += 1

    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

```python
def deep_flatten(lst):
    result = []

    for i in lst:
        if isinstance(i, list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)

    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
def average_nested(lst):

    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

```python
#element wise add
import time
size = 1_000_000

list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result = [a + b for a, b in zip(list1, list2)]
end = time.time()

print("Python List Addition Time:", end - start)
```
